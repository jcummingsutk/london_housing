**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
# I think I can do this with my groupby I did earlier, though i'm going to use 2001 to 2021 instead.
properties_by_year = properties_by_year.set_index(['borough'])
properties
```

```python
# Creating the price ratio dictionary. Look through all boroughs in the index. Then query for the years 2021 and 2001 and get the average price, then compute the ratio.
price_ratio_dict = {}
for brgh in properties_by_year.index.unique():
    price_2001_series = properties_by_year.loc[brgh].query("year == 2001")['average price'] #queries
    price_2021_series = properties_by_year.loc[brgh].query("year == 2021")['average price']
    price_ratio_dict[brgh] = price_2021_series[0]/price_2001_series[0] #get ratio
print(price_ratio_dict) #just checking output
```

```python
# I want to find the 5 maximum ratios and do a bar chart
def get_maxs(in_dict, N=1):
    """ returns the top N values of the in_dict. Assumes there are N values, will throw errors if not """
    brgh_avgs = in_dict.copy()
    counter = 0 # is there a more pythonic way to do this?
    maxs = {}
    while (counter < N):
        max_key = max(brgh_avgs, key=brgh_avgs.get)
        maxs[max_key] = brgh_avgs.pop(max_key)
        counter += 1
    return maxs
```

```python
max_dict = get_maxs(price_ratio_dict, 10)
print(max_dict)
```

```python
keys = list(max_dict.keys())
vals = [float(max_dict[k]) for k in keys]
print(keys)
sns.barplot(x = keys, y = vals)
```
